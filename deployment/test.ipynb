```python
import pandas as pd
import joblib  # or 'import pickle'
import json
import os
```

```python
# Load the model
MODEL_PATH = './models'

lr_model = joblib.load(os.path.join(MODEL_PATH, 'lr.pkl'))
rf_model = joblib.load(os.path.join(MODEL_PATH, 'rf.pkl'))
xgb_model = joblib.load(os.path.join(MODEL_PATH, 'xgb.pkl'))
lstm_model = joblib.load(os.path.join(MODEL_PATH, 'lstm.pkl'))
```

```python
# Load your 3 records
TEST_DATA_PATH = './data'

with open(os.path.join(TEST_DATA_PATH, 'test.json'), 'r') as f:
    data = json.load(f)

# Convert JSON to a DataFrame
# (This ensures column names and order match your training data)
test_df = pd.DataFrame(data)
```

```python
try:
    lr_predictions = lr_model.predict(test_df)  # Get class predictions (0 or 1)
    lr_probs = lr_model.predict_proba(test_df)  # Get probabilities (optional, but helpful for debugging)   

    print("===== LOGISTIC REGRESSION =====")
    print(f"Predictions: {lr_predictions}")
    print(f"Probabilities: \n{lr_probs}")
    print(f"{'='*30}\n")
except Exception as e:
    print(f"Error running Logistic Regression model: {e}")

try:
    rf_predictions = rf_model.predict(test_df)
    rf_probs = rf_model.predict_proba(test_df)

    print("===== RANDOM FOREST =====")
    print(f"Predictions: {rf_predictions}")
    print(f"Probabilities: \n{rf_probs}")
    print(f"{'='*30}\n")
except Exception as e:
    print(f"Error running Random Forest model: {e}")

try:
    xgb_predictions = xgb_model.predict(test_df)
    xgb_probs = xgb_model.predict_proba(test_df)

    print("===== XGBOOST =====")
    print(f"Predictions: {xgb_predictions}")
    print(f"Probabilities: \n{xgb_probs}")
    print(f"{'='*30}\n")
except Exception as e:
    print(f"Error running XGBoost model: {e}")

try:
    lstm_predictions = lstm_model.predict(test_df)
    lstm_probs = lstm_model.predict_proba(test_df)

    print("===== LSTM =====")
    print(f"Predictions: {lstm_predictions}")
    print(f"Probabilities: \n{lstm_probs}")
except Exception as e:
    print(f"Error running LSTM model: {e}")
```
